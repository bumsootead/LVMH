```python
# Volume Prediction Demo

#This notebook loads the regressor saved by `main.ipynb` and lets you input price data to predict trading volume. Be sure the environment has `pandas`, `scikit-learn`, `matplotlib`, and `joblib` installed.

#Model filename changed to `volume_model.pkl` to reflect regression task.
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from sklearn.preprocessing import StandardScaler
import joblib

# Load the trained regression model
volume_model = joblib.load("volume_model.pkl")
```

```python
# 입력 데이터 수집
Open = float(input("Open: "))
High = float(input("High: "))
Low = float(input("Low: "))
Close = float(input("Close: "))

# normalization using the same scaler as training data
# (StandardScaler was used in `main.ipynb` for regression)
data = pd.read_csv('Data/LV.csv')
x_data = data[['Open', 'High', 'Low', 'Close']]
scaler = StandardScaler()
scaler.fit(x_data)

input_data = [[Open, High, Low, Close]]
input_data_normalized = scaler.transform(input_data)

print("Input data:")
print(input_data)
print("\nNormalized data:")
print(input_data_normalized)

# model.predict() using the loaded regressor to predict volume
predicted_volume = volume_model.predict(input_data_normalized)
print("\nPredicted volume:")
print(predicted_volume[0])
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
# Data Load
data = pd.read_csv('Data/LV.csv')

# Date
data['Date'] = pd.to_datetime(data['Date'])
data_sorted = data.sort_values('Date').reset_index(drop=True)

# Graph Creation (4 subplots)
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Stock Price Analysis: Mean and Standard Deviation', fontsize=16)

# Opening Price
ax = axes[0, 0]
mean_open = data_sorted['Open'].mean()
std_open = data_sorted['Open'].std()
ax.errorbar(range(len(data_sorted)), data_sorted['Open'], yerr=std_open, 
            fmt='o-', alpha=0.6, capsize=3, label='Open')
ax.axhline(mean_open, color='r', linestyle='--', label=f'Mean: {mean_open:.2f}')
ax.fill_between(range(len(data_sorted)), mean_open - std_open, mean_open + std_open, 
                alpha=0.2, color='r', label=f'±Std: {std_open:.2f}')
ax.set_title('Opening Prices')
ax.set_xlabel('Date Index')
ax.set_ylabel('Price')
ax.legend()
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
